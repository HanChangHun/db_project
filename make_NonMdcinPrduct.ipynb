```python
output_path = Path("output")
output_file = Path("NonMdcinPrduct.csv")
if not Path.is_dir(output_path):
    Path.mkdir(output_path)
    
# get page number
queryParams = '?' + urlencode({quote_plus('pageNo') : '1', quote_plus('numOfRows') : '1'})
url = 'http://opendata.mfds.go.kr/portal/NonMdcinPrductPrmisnInfoService/getNonMdcinPrductPrmisnInfoList' + queryParams
data = urlopen(url).read()
soup = BeautifulSoup(data, "html.parser")
num_page = int(soup.find('totalcount').text) // 100 + 1

temp_dict = {'품목기준코드' : [], '품목명' : [], '효능효과' : [], '용법용량' : [],
             '주의사항(일반)' : [], '품목코드' : [], '품목코드명' : [], '업체명' : []}
for i in range(1, num_page + 1):
    print("{0} / {1}".format(i, num_page))
    queryParams = '?' + urlencode({quote_plus('pageNo') : '{0}'.format(i), quote_plus('numOfRows') : '100'})
    url = 'http://opendata.mfds.go.kr/portal/NonMdcinPrductPrmisnInfoService/getNonMdcinPrductPrmisnInfoList' + queryParams
    
    data = urlopen(url).read()
    soup = BeautifulSoup(data, "html.parser")
    items = soup.find("items")
    
    for item in items.findAll("item"):
        item_seq = item.item_seq.text.strip()
        item_name = item.item_name.text.strip() if len(item.item_name.text.strip()) < 20000 else "None"
        ee_doc_data = item.ee_doc_data.text.strip() if len(item.ee_doc_data.text.strip()) < 20000 else "None"
        ud_doc_data = item.ud_doc_data.text.strip() if len(item.ud_doc_data.text.strip()) < 20000 else "None"
        nb_doc_data = item.nb_doc_data.text.strip() if len(item.nb_doc_data.text.strip()) < 20000 else "없음"
        class_no = item.class_no.text.strip()
        class_no_name = item.class_no_name.text.strip() if len(item.class_no_name.text.strip()) < 20000 else "None"
        entp_name = item.entp_name.text.strip() if len(item.entp_name.text.strip()) < 20000 else "None"

        if item_seq.isdigit() and class_no.isdigit():
            temp_dict['품목기준코드'].append(item_seq)
            temp_dict['품목명'].append(item_name)
            temp_dict['효능효과'].append(ee_doc_data)
            temp_dict['용법용량'].append(ud_doc_data)
            temp_dict['주의사항(일반)'].append(nb_doc_data)
            temp_dict['품목코드'].append(class_no)
            temp_dict['품목코드명'].append(class_no_name)
            temp_dict['업체명'].append(entp_name)
        else:
            print("{} data is not valid, please check the data".format(item_seq))

df = pd.DataFrame.from_dict(temp_dict)
df.to_csv(Path.joinpath(output_path, output_file), encoding='euc-kr', index=False)
```
